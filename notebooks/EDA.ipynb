# Corr√©lations

üîé S√©lection des colonnes num√©riques uniquement

```python
df_numerique = df.select_dtypes(include=["float64", "int64", "bool"]).copy()
```

üîó Matrice de corr√©lation

```python
correlation_matrix = df_numerique.corr()
```

üìå Afficher la matrice tri√©e par corr√©lation avec HTA

```python
print("üìà Corr√©lations avec la variable HTA :")
print(correlation_matrix["HTA"].sort_values(ascending=False))
```

üî• Heatmap de la matrice de corr√©lation

```python
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True,
            cbar_kws={"shrink": .8}, linewidths=.5)
plt.title("üîó Matrice de corr√©lation (Pearson)")
plt.tight_layout()
plt.show()
```

üí° Option : heatmap centr√©e sur les variables les plus corr√©l√©es avec HTA

```python
top_corr = correlation_matrix["HTA"].abs().sort_values(ascending=False)[1:11].index.tolist()

plt.figure(figsize=(10, 8))
sns.heatmap(df_numerique[top_corr + ["HTA"]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("üîç Variables les plus corr√©l√©es avec HTA")
plt.tight_layout()
plt.show()
```

# Comparaison selon HTA

# Nouvelles variables

üßì Cr√©ation de tranches d‚Äô√¢ge

```python
df["groupe_age"] = pd.cut(df["age_annees"],
                          bins=[0, 18, 30, 45, 60, 80, 120],
                          labels=["<18", "18‚Äì30", "30‚Äì45", "45‚Äì60", "60‚Äì80", "80+"])
```

üßÇ Moyenne sodium jour 1 + jour 2

```python
df["sodium_moyen_mg"] = df[["sodium_jour1_mg", "sodium_jour2_mg"]].mean(axis=1)
```

‚ù§Ô∏è Fr√©quence cardiaque moyenne (si donn√©es disponibles)

```python
if "frequence_cardiaque_1_bpm" in df.columns:
    df["frequence_cardiaque_moy_bpm"] = df[[
        "frequence_cardiaque_1_bpm",
        "frequence_cardiaque_2_bpm",
        "frequence_cardiaque_3_bpm"
    ]].mean(axis=1)
```

üî¥ Cr√©ation d‚Äôun indicateur "HTA isol√©e systolique"

```python
df["hta_isolee_systolique"] = (
    (df["tension_sys_moy_mmHg"] >= 140) & (df["tension_dia_moy_mmHg"] < 90)
)
```

üíæ Sauvegarder le nouveau dataset avec les variables cr√©√©es

```python
df.to_csv("hta_nhanes_enrichi.csv", index=False)
print("‚úÖ Nouvelles variables ajout√©es et fichier sauvegard√© sous : hta_nhanes_enrichi.csv")
```

# Analyses et interpr√©tations finales

Charger le dataset enrichi

```python
df = pd.read_csv("hta_nhanes_enrichi.csv")
```

üìå 1. R√©partition HTA selon les tranches d‚Äô√¢ge

```python
hta_age = pd.crosstab(df["groupe_age"], df["HTA"], normalize="index") * 100
hta_age.plot(kind="bar", stacked=True, figsize=(8, 5), colormap="coolwarm")
plt.title("üìä Pourcentage de personnes hypertendues par tranche d‚Äô√¢ge")
plt.ylabel("% de personnes")
plt.xlabel("Tranche d‚Äô√¢ge")
plt.legend(["Non HTA", "HTA"], title="HTA")
plt.tight_layout()
plt.show()
```

üìå 2. R√©partition HTA selon le sexe

```python
hta_sexe = pd.crosstab(df["sexe_1H_2F"], df["HTA"], normalize="index") * 100
hta_sexe.index = ["Homme", "Femme"]
hta_sexe.plot(kind="bar", stacked=True, figsize=(6, 4), colormap="viridis")
plt.title("üìä Hypertension par sexe")
plt.ylabel("%")
plt.xlabel("Sexe")
plt.legend(["Non HTA", "HTA"], title="HTA")
plt.tight_layout()
plt.show()
```

üìå 3. Variable importante : IMC selon HTA

```python
sns.boxplot(x="HTA", y="imc", data=df, palette="Set2")
plt.title("Distribution de l'IMC selon la pr√©sence d'HTA")
plt.xlabel("HTA")
plt.ylabel("Indice de Masse Corporelle (IMC)")
plt.tight_layout()
plt.show()
```
