```python
# %%
import pandas as pd
df = pd.read_csv("../data/raw/dataset_gad.csv")
df.head()
```

```python
df.columns
```


- **Age** : L'âge de la personne.
- **Sex** : Le sexe biologique de la personne (masculin ou féminin).
- **HighChol** : Indique si la personne a un taux de cholestérol élevé.
- **CholCheck** : Indique si la personne a fait vérifier son cholestérol au cours des dernières années.
- **BMI** : Indice de masse corporelle (Body Mass Index), utilisé pour évaluer la corpulence d’une personne.
- **Smoker** : Indique si la personne est fumeuse ou non.
- **HeartDiseaseorAttack** : Indique si la personne a eu une maladie cardiaque ou une crise cardiaque.
- **PhysActivity** : Indique si la personne pratique une activité physique régulière.
- **Fruits** : Indique la fréquence de consommation de fruits par la personne.
- **Veggies** : Indique la fréquence de consommation de légumes.
- **HvyAlcoholConsump** : Indique si la personne a une consommation excessive d’alcool.
- **GenHlth** : Évaluation de la santé générale de la personne.
- **MentHlth** : Évaluation de la santé mentale de la personne (nombre de jours où elle s'est sentie mal mentalement).
- **PhysHlth** : Évaluation de la santé physique de la personne (nombre de jours où elle s'est sentie mal physiquement).
- **DiffWalk** : Indique si la personne a des difficultés à marcher.
- **Stroke** : Indique si la personne a eu un accident vasculaire cérébral (AVC).
- **HighBP** : Indique si la personne souffre d'hypertension artérielle.
- **Diabetes** : Indique si la personne est atteinte de diabète.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Répartition homme/femme
sns.countplot(x='Sex', data=df)
plt.title('Répartition des sexes')
plt.show()

# Nombre de fumeurs
sns.countplot(x='Smoker', data=df)
plt.title('Répartition des fumeurs')
plt.show()
```

```python
# Boxplot du BMI pour détecter les valeurs aberrantes
sns.boxplot(x=df['BMI'])
plt.title('Distribution du BMI')
plt.show()
```

```python
import numpy as np

# Matrice de corrélation
corr_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matrice de corrélation')
plt.show()
```
